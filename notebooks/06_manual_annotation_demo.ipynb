# 06 — Manual Annotation Demo

Demonstrates the annotation workflow:
1. Load the annotation template
2. Validate demo annotation records
3. Merge new annotations with existing records
4. Show validation error reporting

```python
from __future__ import annotations

import pathlib
import pandas as pd
from gdynia_thermal_audit.annotation.templates import get_annotation_template, create_annotation_csv
from gdynia_thermal_audit.annotation.validate import validate_annotations
from gdynia_thermal_audit.annotation.merge import merge_annotations
from gdynia_thermal_audit.utils.io import ensure_dir

ensure_dir('data/annotations')
```

```python
# Show template structure
df_template = get_annotation_template()
print('Annotation template columns:')
for col in df_template.columns:
    print(f'  {col}')
```

```python
# Load demo annotations and validate
df_demo = pd.read_csv('data/demo/demo_annotations.csv')
is_valid, errors = validate_annotations(df_demo)
print(f'Valid: {is_valid}')
if errors:
    print(f'{len(errors)} validation issue(s):')
    for e in errors:
        print(f'  - {e}')
else:
    print('All records passed validation')
```

```python
# Simulate merging new annotations into existing set
new_record = df_demo.iloc[:2].copy()
new_record['record_id'] = ['ANN-99990101-0099', 'ANN-99990101-0100']
merged = merge_annotations(df_demo, new_record)
print(f'Before merge: {len(df_demo)} records')
print(f'After merge:  {len(merged)} records')
```

```python
# Create a blank template CSV for real annotations
create_annotation_csv('data/annotations/annotation_template.csv', n_rows=0)
print('Blank template saved to data/annotations/annotation_template.csv')
```
